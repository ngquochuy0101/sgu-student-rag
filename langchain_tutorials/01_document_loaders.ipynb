# 1. Document Loaders (Nhập liệu)

## 📖 Lý thuyết
Trong LangChain, **Document Loaders** (Trình tải tài liệu) được sử dụng để tải dữ liệu từ nhiều nguồn khác nhau (như file PDF, text, database, trang web) và chuyển đổi chúng thành định dạng `Document` thống nhất.
Một `Document` gồm 2 thành phần chính:
- `page_content`: Nội dung văn bản (text).
- `metadata`: Siêu dữ liệu đi kèm (ví dụ: tên file, số trang, tác giả).

Trong dự án RAG SGU, bạn đang sử dụng kết hợp `DirectoryLoader` (để duyệt qua thư mục) và `PyPDFLoader` (để phân tích từng file PDF).

## 💻 Code mẫu
Ví dụ tối giản: tạo một đoạn text ngắn và chuyển thành `Document` (không đọc PDF).

```python
import unicodedata
import re

def clean_text(text: str) -> str:
    """
    Hàm làm sạch văn bản:
    1. Chuẩn hóa font chữ Unicode (NFC).
    2. Loại bỏ các khoảng trắng, dấu tab, dấu xuống dòng thừa.
    """
    normalized = unicodedata.normalize("NFC", text or "")
    # Thay thế 1 hoặc nhiều khoảng trắng (bao gồm cả \n, \t) bằng 1 khoảng trắng duy nhất
    return re.sub(r"\s+", " ", normalized).strip()

# 1. Tạo một đoạn text nhỏ để test (cố tình để lộn xộn)
sample_text = """
    Đây là   một đoạn text     nhỏ để test.
    
    Nó có chứa rất nhiều khoảng trắng thừa
    và      dấu xuống dòng không cần thiết.
"""

# 2. In ra đoạn text ban đầu
print("--- Văn bản trước khi làm sạch ---")
# Dùng hàm repr() để bạn có thể nhìn thấy rõ các ký tự ẩn như '\n' (xuống dòng)
print(repr(sample_text)) 

print("\n" + "="*40 + "\n")

# 3. Chạy hàm làm sạch và in ra kết quả
cleaned_text = clean_text(sample_text)
print("--- Văn bản sau khi làm sạch ---")
print(repr(cleaned_text))
print("\nHiển thị bình thường:")
print(cleaned_text)
```

## 🎯 Bài tập nhỏ
**Yêu cầu**: Tạo 3 đoạn text ngắn, chuyển thành danh sách `Document`, rồi in ra tổng số `Document` và 1 metadata mẫu.

```python
from langchain_core.documents import Document

# 1. Tạo 3 đoạn text ngắn
texts = [
    "Python là một ngôn ngữ lập trình mạnh mẽ và dễ học.",
    "LangChain giúp xây dựng các ứng dụng ngôn ngữ lớn (LLMs) dễ dàng hơn.",
    "RAG (Retrieval-Augmented Generation) là kỹ thuật giúp AI trả lời chính xác dựa trên tài liệu riêng."
]

# 2. Chuyển thành danh sách Document
documents = []
for index, text in enumerate(texts):
    # Tạo metadata đơn giản cho từng đoạn text
    # (Ví dụ: lưu số thứ tự và phân loại nguồn)
    doc_metadata = {
        "id": index + 1, 
        "source": "bai_tap_tao_document",
        "category": "education"
    }
    
    # Khởi tạo đối tượng Document gồm nội dung (page_content) và siêu dữ liệu (metadata)
    doc = Document(page_content=text, metadata=doc_metadata)
    
    # Thêm vào danh sách
    documents.append(doc)

# 3. In ra tổng số Document
print(f"Tổng số Document hiện có: {len(documents)}\n")

# 4. In ra 1 metadata mẫu (của Document đầu tiên)
if documents: # Kiểm tra xem danh sách có phần tử nào không
    first_doc = documents[0]
    print("--- Thông tin Document mẫu ---")
    print(f"Metadata : {first_doc.metadata}")
    print(f"Nội dung : {first_doc.page_content}")
```
