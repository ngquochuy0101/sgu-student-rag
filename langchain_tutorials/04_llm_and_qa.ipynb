# 4. LLM & QA Retrieval (Prompt & Hỏi đáp)

## 📖 Lý thuyết
Giai đoạn cuối cùng của RAG là **Generation**. Ở bước này, chúng ta sẽ thiết kế một 
**PromptTemplate** để ép ngôn ngữ tự nhiên thành một cấu trúc: đưa "Câu hỏi" và "Ngữ cảnh được truy xuất (Context)" vào cho LLM đọc và tổng hợp câu trả lời.

Dự án sử dụng mô hình Google Gemini (`ChatGoogleGenerativeAI`).

## 💻 Code mẫu
Sử dụng Gemini để sinh câu trả lời dựa trên context (phần mô phỏng giống hàm `_generate_answer` trong `qa_service.py`).

```python
import os
import sys
sys.path.append(os.path.abspath('..'))
from dotenv import load_dotenv

# Load biến môi trường từ file .env
load_dotenv("../.env")

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate

# 1. Khởi tạo LLM Gemini
api_key = os.getenv("GOOGLE_API_KEY")
if not api_key:
    print("Lỗi: Không tìm thấy GOOGLE_API_KEY trong môi trường.")
else:
    print("Đã load API Key thành công.")
    llm = ChatGoogleGenerativeAI(
        model="gemini-2.5-flash",
        google_api_key=api_key,
        temperature=0.5
    )
    
    # 2. Xây dựng Prompt (System Prompt)
    prompt_text = """Bạn là trợ lý giải đáp thắc mắc.
    
    THÔNG TIN TỪ TÀI LIỆU:
    {context}
    
    CÂU HỎI: {question}
    
    YÊU CẦU: Trả lời chính xác dựa theo tài liệu. Nếu không biết thì nói "Không có thông tin".
    TRẢ LỜI:
    """
    prompt = PromptTemplate.from_template(prompt_text)
    
    # 3. Tạo dữ liệu giả định
    sample_context = "Năm học tại Đại học Sài Gòn bắt đầu từ tháng 9 và chia thành 3 học kỳ (kể cả Hè)."
    sample_question = "Năm học có mấy học kỳ?"
    
    # 4. Truyền dữ liệu vào Prompt và kích hoạt LLM
    final_prompt = prompt.format(context=sample_context, question=sample_question)
    print("\n--- PROMPT GỬI CHO LLM ---")
    print(final_prompt)
    
    print("\n--- CÂU TRẢ LỜI CỦA GEMINI ---")
    # Thực thi invokes LLM (Cần có mã API hợp lệ)
    response = llm.invoke(final_prompt) 
    print(response.content)
```

## 🎯 Bài tập nhỏ
**Yêu cầu**: Hãy đổi `prompt_text` để yêu cầu AI "luôn luôn trả lời bằng ngôn ngữ Tiếng Anh". Sau đó truyền vào mẫu câu hỏi Tiếng Việt và xem AI đáp lại bằng Tiếng Anh.

```python
if 'llm' in locals() and llm is not None:
    # 1. Đổi PromptTemplate sang yêu cầu Tiếng Anh
    english_prompt_text = """You are a helpful AI assistant.
    
    INFORMATION:
    {context}
    
    QUESTION: {question}
    
    REQUIREMENT: You MUST answer the question in English completely natively, based ONLY on the provided information.
    ANSWER:
    """
    english_prompt = PromptTemplate.from_template(english_prompt_text)
    
    final_english_prompt = english_prompt.format(
        context="Năm học tại Đại học Sài Gòn bắt đầu từ tháng 9 và chia thành 3 học kỳ (kể cả Hè).", 
        question="Năm học có mấy học kỳ?"
    )
    print("\n--- PROMPT TIẾNG ANH GỬI CHO LLM ---")
    print(final_english_prompt)
    
    print("\n--- CÂU TRẢ LỜI CỦA GEMINI ---")
    try:
        # 2. Invoke gọi lại api
        response = llm.invoke(final_english_prompt)
        print(response.content)
    except Exception as e:
        print("Lỗi gọi API:", e)
else:
    print("Bạn cần khởi tạo biến 'llm' phía trên (Cần có GOOGLE_API_KEY) để tiến hành bài tập này.")
```
